```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import seaborn as sns
%matplotlib inline
```

```python
dfiris=pd.read_csv('iris.csv', sep=',', encoding='utf-8')
```

```python
dfiris
```

```python
sns.pairplot(dfiris,hue="species");
```

```python
sns.lmplot(x="sepal_length", y="sepal_width", hue="species",
               truncate=True, size=5, data=dfiris)
```

```python
sns.lmplot(x="petal_length", y="petal_width", hue="species",
               truncate=True, size=5, data=dfiris)
```

```python
dfiris.columns
```

```python
x=dfiris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
```

```python
y=dfiris[['species']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
irismodel = LogisticRegression()
```

```python
irismodel.fit(x_train,y_train)
```

```python
predito = irismodel.predict(x_test)
```
